# VS Code Python 一体化配置实战

> 本 Notebook 演示了如何自动生成和验证项目的 VS Code 配置文件。

## 1. 自动生成 settings.json
将推荐配置导出为项目内的 JSON 文件。

```python
import json
config = {
    "editor.formatOnSave": True,
    "python.analysis.typeCheckingMode": "basic",
    "editor.defaultFormatter": "ms-python.black-formatter"
}
print(json.dumps(config, indent=2))
```

## 2. 检查团队合规性
编写一段代码检查当前工作区的设置是否满足最低质量要求。

```python
def check_compliance(settings_dict):
    required = {"editor.formatOnSave": True}
    for key, expected in required.items():
        actual = settings_dict.get(key)
        if actual == expected:
            print(f"[OK] {key} is correct.")
        else:
            print(f"[!!] {key} mismatch! Expected {expected}, got {actual}")

check_compliance(config)
```
