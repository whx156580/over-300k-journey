## 1. Python Doctor (环境诊断)
一键输出当前解释器与系统的核心属性。

```python
import sys
import platform
from pathlib import Path

report = {
    "Version": sys.version.split()[0],
    "Platform": platform.system(),
    "In Venv": sys.prefix != getattr(sys, 'base_prefix', sys.prefix),
    "CWD": Path.cwd().name
}

for k, v in report.items():
    print(f"{k:<12}: {v}")
```

## 2. 自动化配置 (Settings Gen)
模拟初始化项目所需的 VS Code 设置。

```python
import json
settings = {
    "python.analysis.typeCheckingMode": "basic",
    "editor.formatOnSave": True
}
print("Recommended settings.json content:")
print(json.dumps(settings, indent=4))
```
