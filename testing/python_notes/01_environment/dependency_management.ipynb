# Python 依赖管理 (Dependency Management) 实战

> 本 Notebook 展示了如何使用 `importlib.metadata` 审计环境依赖，并解析常见的依赖描述文件。

## 1. 自动化环境审计
检查当前运行环境是否满足项目要求的包版本。

```python
import importlib.metadata

def audit_env(packages):
    for pkg in packages:
        try:
            version = importlib.metadata.version(pkg)
            print(f"[OK] {pkg:<15} : {version}")
        except importlib.metadata.PackageNotFoundError:
            print(f"[!!] {pkg:<15} : NOT INSTALLED")

audit_env(["pytest", "requests", "pandas", "non_existent"])
```

## 2. 解析 requirements.txt
演示如何清洗原始文本，提取可用的依赖清单。

```python
raw_content = """
pytest>=8.0.0
# CI tools
flake8==7.0.0
requests  # Latest version
"""

cleaned = [line.split('#')[0].strip() for line in raw_content.splitlines() if line.strip() and not line.strip().startswith('#')]
print(f"Cleaned requirements: {cleaned}")
```

## 3. 生成 pyproject.toml 映射
将包列表转换为 Poetry 格式。

```python
pkgs = [("black", "24.1.0"), ("mypy", "1.8.0")]
for p, v in pkgs:
    print(f'{p} = "^{v}"')
```
