## 1. 技能矩阵建模
将抽象的岗位需求转化为结构化的字典数据。

```python
matrix = {
    "SDET": ["pytest", "playwright", "fastapi"],
    "Backend": ["Django", "Postgres", "Redis"]
}
for role, skills in matrix.items():
    print(f"Role: {role} | Top Skills: {', '.join(skills)}")
```

## 2. 学习优先级决策
简单的量化算法，帮助确定下一个学习专题。

```python
def priority(interest, demand):
    return "CRITICAL" if (interest + demand) > 1.5 else "LATER"

print(f"Next Topic Priority: {priority(0.9, 0.8)}")
```
