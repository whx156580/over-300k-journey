# Python 多版本管理与虚拟环境实战

> 本 Notebook 展示了如何通过 Python 代码自检解释器版本、探测虚拟环境状态以及管理本地路径。

## 1. 解释器详情自检
输出当前内核的所有关键环境属性。

```python
import sys
import os
print(f"Python Version: {sys.version}")
print(f"Executable: {sys.executable}")
print(f"Prefix: {sys.prefix}")
print(f"Platform: {sys.platform}")
```

## 2. 虚拟环境探测 (PEP 405)
逻辑：如果 `sys.prefix` 与 `sys.base_prefix` 不同，则处于虚拟环境中。

```python
is_venv = sys.prefix != getattr(sys, 'base_prefix', sys.prefix)
status = "[ISOLATED]" if is_venv else "[SYSTEM/GLOBAL]"
print(f"Environment Status: {status}")
```

## 3. 路径搜索链 (sys.path)
查看 Python 寻找模块的优先级顺序。

```python
for i, p in enumerate(sys.path):
    print(f"{i}: {p}")
```
