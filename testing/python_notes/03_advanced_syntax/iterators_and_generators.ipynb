# 迭代器与生成器 (Iterators & Generators) 实战

> 本 Notebook 展示了如何利用 Python 的惰性求值特性处理大数据流，并演示生成器的双向通信能力。

## 1. 无限数据流水线
使用 `itertools` 配合生成器实现高效的流式处理。

```python
import itertools

def data_source():
    for i in itertools.count(1):
        yield i

def processor(stream):
    for val in stream:
        if val % 3 == 0:
            yield f"Processed: {val}"

pipeline = processor(data_source())
print(list(itertools.islice(pipeline, 5)))
```

## 2. 双向通信累加器
利用 `.send()` 方法在运行时修改生成器内部变量。

```python
def accumulator():
    total = 0
    while True:
        val = yield total
        if val is not None:
            total += val

gen = accumulator()
print(f"Start: {next(gen)}")
print(f"Add 5: {gen.send(5)}")
print(f"Add 10: {gen.send(10)}")
```

## 3. 内存开销对比
直观展示 `list` 与 `generator` 的内存占用差异。

```python
import sys
n = 1_000_000
l = [i for i in range(n)]
g = (i for i in range(n))
print(f"List size: {sys.getsizeof(l) / 1024 / 1024:.2f} MB")
print(f"Generator size: {sys.getsizeof(g)} Bytes")
```
