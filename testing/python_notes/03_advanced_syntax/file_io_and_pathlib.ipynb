# 文件读写与路径进阶实战 (File I/O & Pathlib)

> 本 Notebook 展示了如何使用 `pathlib` 处理跨平台路径，以及如何进行高效的流式文件读取。

## 1. 面向对象的路径操作
演示 `Path` 的拼接、父目录回溯及属性检查。

```python
from pathlib import Path

p = Path("some/sub/folder/file.txt")
print(f"Name: {p.name}")
print(f"Suffix: {p.suffix}")
print(f"Parent: {p.parent}")
print(f"Cross-platform: {p}")
```

## 2. 文本读写最佳实践
使用 `write_text` 和 `read_text` 简化小型文件操作。

```python
temp = Path("demo_output.txt")
temp.write_text("First Line\nSecond Line", encoding="utf-8")
print(f"Read back:\n{temp.read_text(encoding='utf-8')}")
temp.unlink() # 删除临时文件
```

## 3. 流式读取大数据
演示逐行迭代以保持低内存占用。

```python
with open("README.md", "r", encoding="utf-8") as f:
    # 仅读取前 3 行
    for i, line in enumerate(f):
        if i >= 3: break
        print(f"L{i+1}: {line.strip()}")
```
