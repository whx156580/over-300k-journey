## 1. 结构化日志扫描
演示如何使用生成器边解析边过滤，保持低内存占用。

```python
import re
from dataclasses import dataclass

@dataclass
class Entry:
    lvl: str
    msg: str

def scan(lines):
    for line in lines:
        if m := re.match(r"\[(\w+)\] (.*)", line):
            yield Entry(m.group(1), m.group(2))

raw = ["[INFO] Start", "[DEBUG] Init", "[ERROR] Fail"]
for item in scan(raw):
    if item.lvl == "ERROR":
        print(f"ALERT: {item.msg}")
```

## 2. 磁盘资源审计
利用 `pathlib` 的面向对象接口快速盘点文件。

```python
from pathlib import Path
root = Path(".")
for ext in [".md", ".py", ".ipynb"]:
    count = len(list(root.glob(f"*{ext}")))
    print(f"{ext:<7}: {count} files")
```
