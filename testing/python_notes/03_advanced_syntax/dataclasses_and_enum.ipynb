# 结构化建模实战 (Dataclasses & Enum)

> 本 Notebook 展示了如何使用 `dataclass` 消除样板代码，并结合 `Enum` 实现类型安全的状态管理。

## 1. 自动生成方法演示
观察 `@dataclass` 如何自动提供可读的 `__repr__` 和 `__eq__`。

```python
from dataclasses import dataclass

@dataclass
class User:
    id: int
    name: str

u1 = User(1, "Alice")
u2 = User(1, "Alice")
print(f"Repr: {u1}")
print(f"Equal? {u1 == u2}")
```

## 2. 默认值工厂 (Default Factory)
演示如何安全地初始化可变字段。

```python
from dataclasses import field

@dataclass
class Group:
    members: list = field(default_factory=list)

g1, g2 = Group(), Group()
g1.members.append("Alice")
print(f"G1: {g1.members}, G2: {g2.members}")
```

## 3. 强类型枚举 (Enum)
替代不稳定的字符串常量。

```python
from enum import Enum
class Color(Enum):
    RED = 1
    GREEN = 2

print(f"Selected: {Color.RED.name} -> {Color.RED.value}")
```
