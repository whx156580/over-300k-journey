# 断点调试与 PDB 进阶实战

> 本 Notebook 展示了如何利用 Python 3.7+ 的 `breakpoint()` 机制进行高效调试。

## 1. 自动 Hook 演示
将原本阻塞的 `pdb` 交互重定向为非阻塞的日志输出。

```python
import sys
def my_hook(*args, **kwargs):
    print("!!! BREAKPOINT TRIGGERED !!!")

sys.breakpointhook = my_hook
x = 100
breakpoint()
print("Done")
```
