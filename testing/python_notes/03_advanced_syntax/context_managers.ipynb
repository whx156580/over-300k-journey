# 上下文管理器进阶实战 (Context Managers)

> 本 Notebook 展示了如何利用 `with` 语句实现资源的自动清理和状态的临时切换。

## 1. 资源管理器协议
演示自定义类实现 `__enter__` 与 `__exit__`。

```python
class FileLogger:
    def __enter__(self):
        print("Log session started")
        return self
    def __exit__(self, *args):
        print("Log session ended")

with FileLogger():
    print("Writing logs...")
```

## 2. 状态切换装饰器
使用 `contextlib.contextmanager` 简化代码。

```python
from contextlib import contextmanager

@contextmanager
def tags(name):
    print(f"<{name}>", end="")
    yield
    print(f"</{name}>")

with tags("h1"): print("Hello World", end="")
```

## 3. 性能测试：with vs try-finally
观察两种写法在性能上的微小差异。

```python
import timeit
def use_with():
    with open("README.md", "r") as f: pass

def use_try():
    f = open("README.md", "r")
    try: pass
    finally: f.close()

print(f"With: {timeit.timeit(use_with, number=1000):.4f}s")
print(f"Try:  {timeit.timeit(use_try, number=1000):.4f}s")
```
