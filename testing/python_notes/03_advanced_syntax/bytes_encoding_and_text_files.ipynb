# 文本编码与字节流实战 (Encoding & Bytes)

> 本 Notebook 展示了 Python 中文本 (`str`) 与字节 (`bytes`) 的转换关系，并演示了如何安全处理多种编码。

## 1. 编码与解码基础
演示 Unicode 字符在不同编码下的字节表现。

```python
text = "Python 笔记"
utf8_bytes = text.encode("utf-8")
gbk_bytes = text.encode("gbk")

print(f"Original: {text}")
print(f"UTF-8: {utf8_bytes.hex(' ')}")
print(f"GBK:   {gbk_bytes.hex(' ')}")
```

## 2. 字节流长度 vs 字符长度
理解为什么计算文件大小时不能直接用 `len(str)`。

```python
s = "你好!"
b = s.encode("utf-8")
print(f"Char length: {len(s)}")
print(f"Byte length (UTF-8): {len(b)}")
```

## 3. 内存二进制流 (BytesIO)
演示如何在不写磁盘的情况下模拟文件操作。

```python
import io
bio = io.BytesIO(b"\x00\x01\x02\x03")
print(f"First 2 bytes: {bio.read(2).hex()}")
```
