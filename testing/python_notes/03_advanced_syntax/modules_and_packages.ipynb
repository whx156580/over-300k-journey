# 模块与包进阶实战 (Modules & Packages)

> 本 Notebook 展示了 Python 导入系统的缓存机制以及如何使用 `importlib` 进行动态插件加载。

## 1. 动态加载插件
利用 `importlib` 在运行时加载代码，实现系统解耦。

```python
import importlib

def run_math_func(name, val):
    math_mod = importlib.import_module("math")
    func = getattr(math_mod, name)
    return func(val)

print(f"Sqrt of 100: {run_math_func('sqrt', 100)}")
```

## 2. 检查模块缓存
理解 `sys.modules` 如何存储已加载的模块对象。

```python
import sys
import os

print(f"Is 'os' in sys.modules? {'os' in sys.modules}")
print(f"Object ID of 'os': {id(sys.modules['os'])}")
```
