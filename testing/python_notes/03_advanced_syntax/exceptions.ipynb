# 异常体系 (Exception Handling) 实战

> 本 Notebook 展示了 Python 中的异常链、自定义异常以及如何编写异常断言工具。

## 1. 异常链 (raise from)
显式保留原始报错上下文。

```python
class MyError(Exception): pass

try:
    1/0
except ZeroDivisionError as e:
    raise MyError("Calculation failed") from e
```

## 2. 自定义断言上下文管理器
模仿 `pytest.raises` 实现自己的错误捕获器。

```python
import contextlib

@contextlib.contextmanager
def expect_fail(exc_type):
    try:
        yield
    except exc_type:
        print(f"OK: Caught {exc_type.__name__}")
    else:
        raise AssertionError("Failed to catch error")

with expect_fail(TypeError):
    len(123)
```
