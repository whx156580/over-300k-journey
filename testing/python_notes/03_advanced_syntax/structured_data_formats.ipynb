# 序列化与结构化数据实战 (JSON & CSV)

> 本 Notebook 展示了如何处理复杂类型的 JSON 序列化以及高效的 CSV 数据转换。

## 1. 扩展 JSONEncoder
让原生 `json` 模块支持 `datetime` 对象。

```python
import json
from datetime import datetime

class MyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, datetime):
            return obj.strftime('%Y-%m-%d %H:%M:%S')
        return super().default(obj)

data = {"now": datetime.now()}
print(json.dumps(data, cls=MyEncoder))
```

## 2. CSV 数据清洗与转换
利用 `DictReader` 快速处理表格数据。

```python
import csv
import io

csv_raw = "id,user,score\n1,Alice,90\n2,Bob,85"
f = io.StringIO(csv_raw)
reader = csv.DictReader(f)

high_scores = [row for row in reader if int(row['score']) > 88]
print(f"High scores: {high_scores}")
```

## 3. 字典合并 (Python 3.9+)
演示在处理配置数据时常用的合并操作。

```python
default_config = {"host": "localhost", "port": 8000}
user_config = {"port": 9000, "debug": True}

merged = default_config | user_config
print(f"Merged config: {merged}")
```
