# 函数式编程与递归实战 (Functional & Recursion)

> 本 Notebook 展示了 Python 中的高阶函数应用、`partial` 工厂模式以及递归性能优化。

## 1. 匿名函数 Lambda 与排序
演示如何使用 `key` 参数进行动态排序。

```python
data = ["apple", "banana", "kiwi"]
sorted_by_len = sorted(data, key=lambda x: len(x))
print(f"Sorted by length: {sorted_by_len}")
```

## 2. 偏函数 Partial
固定函数参数，创建新的可重用组件。

```python
from functools import partial
def power(base, exp): return base ** exp
square = partial(power, exp=2)
cube = partial(power, exp=3)

print(f"Square of 5: {square(5)}")
print(f"Cube of 5: {cube(5)}")
```

## 3. 递归与缓存优化
演示 `lru_cache` 如何拯救递归性能。

```python
from functools import lru_cache

@lru_cache(maxsize=128)
def fib(n):
    if n < 2: return n
    return fib(n-1) + fib(n-2)

print(f"Fib(100): {fib(100)}")
print(fib.cache_info())
```
