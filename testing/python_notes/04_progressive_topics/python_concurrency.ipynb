# 并发模型 (Concurrency Models) 实战

> 本 Notebook 对比了 Python 中的线程、进程和协程，展示了在不同场景下的选型与实现。

## 1. 线程安全与锁 (Threading + Lock)
展示多线程并发访问共享资源时的竞争问题。

```python
import threading
from concurrent.futures import ThreadPoolExecutor

class Counter:
    def __init__(self):
        self.value = 0
        self.lock = threading.Lock()
    
    def inc(self):
        with self.lock:
            self.value += 1

c = Counter()
with ThreadPoolExecutor(max_workers=5) as ex:
    for _ in range(100): ex.submit(c.inc)
print(f"Counter: {c.value}")
```

## 2. 协程并发 (Asyncio)
在单线程内高效处理大量等待任务。

```python
import asyncio

async def task(id):
    await asyncio.sleep(0.1)
    return f"Task {id} done"

async def run():
    res = await asyncio.gather(*(task(i) for i in range(5)))
    print(res)

# 注意：在 Notebook 中直接使用 await 或 run
await run()
```
