## 1. 自动化质量门禁
演示如何编写脚本自动化判断“是否可发布”。

```python
def can_release(results):
    gates = {
        "Unit Tests": results['ut_pass'],
        "Coverage": results['cov'] >= 90,
        "Lint": results['lint_errors'] == 0
    }
    passed = all(gates.values())
    print(f"Final Decision: {'SHIP IT' if passed else 'BLOCK'}")
    for name, ok in gates.items():
        print(f"- {name}: {'[OK]' if ok else '[FAIL]'}")

can_release({'ut_pass': True, 'cov': 85, 'lint_errors': 0})
```

## 2. 性能预算 (Service Level Objectives)
在交付时量化“快”与“慢”的边界。

```python
actual_p95 = 280
budget_p95 = 300

if actual_p95 < budget_p95:
    print(f"Success: {actual_p95}ms is within budget ({budget_p95}ms)")
else:
    print(f"Warning: Performance degradation detected!")
```
