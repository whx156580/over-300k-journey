# 内存管理与 GC 调优 (Memory Management & GC)

> 本 Notebook 展示了 Python 的引用计数机制、循环引用问题以及如何使用 `gc` 模块进行手动控制。

## 1. 引用计数实验
观察变量赋值如何改变对象的引用计数。

```python
import sys
a = [1, 2, 3]
print(f"Initial: {sys.getrefcount(a)}")
b = a
print(f"After b = a: {sys.getrefcount(a)}")
del b
print(f"After del b: {sys.getrefcount(a)}")
```

## 2. 循环引用与分代回收
模拟循环引用并演示手动 GC 的效果。

```python
import gc

class Node:
    def __init__(self, name): self.name = name
    def __repr__(self): return f"Node({self.name})"

n1 = Node("A")
n2 = Node("B")
n1.next = n2
n2.next = n1

del n1, n2
unreachable = gc.collect()
print(f"Cleaned {unreachable} unreachable objects")
```
