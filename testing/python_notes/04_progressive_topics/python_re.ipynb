# 正则表达式进阶 (Regular Expressions)

> 本 Notebook 展示了 Python `re` 模块的高阶用法，包括命名分组、非贪婪匹配和预编译优化。

## 1. 命名分组：让提取结果具备语义
使用 `(?P<name>...)` 替代脆弱的索引访问。

```python
import re
log = "[ERROR] user=admin action=delete"
pattern = re.compile(r"\[(?P<level>\w+)\]\s+user=(?P<user>\w+)")
match = pattern.search(log)
if match:
    print(f"Dict: {match.groupdict()}")
    print(f"Level: {match.group('level')}")
```

## 2. 非贪婪匹配：精准提取标签
使用 `.*?` 解决“过度吞噬”问题。

```python
html = "<span>One</span><span>Two</span>"
print(f"Greedy: {re.findall(r'<span>(.*)</span>', html)}")
print(f"Non-Greedy: {re.findall(r'<span>(.*?)</span>', html)}")
```

## 3. 性能基准：预编译的收益
对比即时编译与预编译的开销。

```python
import timeit
import re

p_str = r"\d+"
data = "123 456 789"

def raw(): re.findall(p_str, data)
compiled = re.compile(p_str)
def optimized(): compiled.findall(data)

t1 = timeit.timeit(raw, number=100000)
t2 = timeit.timeit(optimized, number=100000)
print(f"Raw: {t1:.4f}s, Optimized: {t2:.4f}s")
```
