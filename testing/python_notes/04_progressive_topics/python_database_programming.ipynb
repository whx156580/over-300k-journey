## 1. 参数化查询：防御 SQL 注入
演示如何使用 `?` 占位符安全地传递参数。

```python
import sqlite3
with sqlite3.connect(":memory:") as conn:
    conn.execute("CREATE TABLE users (id INT, name TEXT)")
    conn.execute("INSERT INTO users VALUES (1, 'Alice')")
    
    # 安全查询
    user_id = 1
    cursor = conn.execute("SELECT * FROM users WHERE id = ?", (user_id,))
    print(f"Found: {cursor.fetchone()}")
```

## 2. 批量写入性能优化
对比 `execute` 与 `executemany` 的巨大差异。

```python
import timeit
data = [(i,) for i in range(1000)]

def slow():
    with sqlite3.connect(":memory:") as c:
        c.execute("CREATE TABLE t (v INT)")
        for d in data: c.execute("INSERT INTO t VALUES (?)", d)

def fast():
    with sqlite3.connect(":memory:") as c:
        c.execute("CREATE TABLE t (v INT)")
        c.executemany("INSERT INTO t VALUES (?)", data)

print(f"Slow: {timeit.timeit(slow, number=10):.4f}s")
print(f"Fast: {timeit.timeit(fast, number=10):.4f}s")
```
