# 变量可变性与深浅拷贝 (Mutability & Copying)

> 本 Notebook 展示了 Python 的对象引用模型，并对比了赋值、浅拷贝与深拷贝的区别。

## 1. 赋值与别名
演示 `b = a` 并不创建新对象。

```python
a = [1, 2, 3]
b = a
print(f"id(a): {id(a)}, id(b): {id(b)}")
print(f"a is b: {a is b}")

b.append(99)
print(f"a after b modified: {a}")
```

## 2. 浅拷贝 (copy.copy)
仅复制一层容器。

```python
import copy
old = [[1, 2], 3]
new = copy.copy(old)

print(f"Outer objects same? {old is new}")
print(f"Inner lists same? {old[0] is new[0]}")

new[0].append(99)
print(f"old after new's inner modified: {old}")
```

## 3. 深拷贝 (copy.deepcopy)
递归复制所有层级。

```python
old = [[1, 2], 3]
new = copy.deepcopy(old)

print(f"Inner lists same? {old[0] is new[0]}")
new[0].append(99)
print(f"old after deep modified: {old}")
```
