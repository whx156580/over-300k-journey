# 运算符与表达式实战 (Operators & Expressions)

> 本 Notebook 展示了 Python 中的链式比较、海象运算符以及短路逻辑的高级技巧。

## 1. 海象运算符 (Walrus Operator)
在表达式内完成赋值，减少重复计算。

```python
import random
def get_data(): return random.randint(0, 100)

if (val := get_data()) > 50:
    print(f"Large value: {val}")
else:
    print(f"Small value: {val}")
```

## 2. 链式比较
优雅地检查数值区间。

```python
age = 25
print(f"Is adult? {18 <= age < 60}")
```

## 3. 短路逻辑与默认值
利用 `or` 运算符提供后备方案。

```python
def fetch_api_key(): return "" # 模拟获取失败

key = fetch_api_key() or "DEFAULT_MOCK_KEY"
print(f"Active Key: {key}")
```
