# 核心容器实战 (Collections)

> 本 Notebook 展示了列表、元组、集合与字典的高级操作及性能对比。

## 1. 集合运算：差异比对
利用集合快速计算并集、交集与差集。

```python
set_a = {"user_id", "name", "email"}
set_b = {"user_id", "last_login", "ip"}

print(f"Common keys: {set_a & set_b}")
print(f"Unique to A: {set_a - set_b}")
print(f"All unique keys: {set_a | set_b}")
```

## 2. 字典合并与默认值
演示 Python 3.9+ 的合并语法及 `defaultdict`。

```python
d1 = {"a": 1, "b": 2}
d2 = {"b": 3, "c": 4}
merged = d1 | d2
print(f"Merged dict: {merged}")

from collections import defaultdict
counter = defaultdict(int)
for char in "mississippi":
    counter[char] += 1
print(f"Counts: {dict(counter)}")
```

## 3. 性能测试：List vs Set
直观对比 $O(n)$ 与 $O(1)$ 的查找开销。

```python
import timeit
n = 10000
l = list(range(n))
s = set(range(n))

t_list = timeit.timeit("9999 in l", globals=locals(), number=1000)
t_set = timeit.timeit("9999 in s", globals=locals(), number=1000)

print(f"List search time: {t_list:.6f}s")
print(f"Set search time: {t_set:.6f}s")
```
