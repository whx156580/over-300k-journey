# 变量与数据类型 (Variables & Data Types)

> 本 Notebook 展示了 Python 的动态强类型特性、变量引用模型以及类型注解的用法。

## 1. 变量是对象的标签
演示变量名只是指向对象的引用。

```python
a = [1, 2, 3]
b = a
print(f"a: {a}, b: {b}")
print(f"Same object? {a is b}")

a = [4, 5]
print(f"After a reassigned - a: {a}, b: {b}")
print(f"Still same object? {a is b}")
```

## 2. 强类型实验
Python 不会隐式转换类型。

```python
try:
    print("Answer: " + 42)
except TypeError as e:
    print(f"TypeError caught: {e}")
    print("Answer: " + str(42))
```

## 3. 类型注解 (Type Hints)
使用注解提升代码可读性。

```python
from typing import List

def greet_all(names: List[str]) -> None:
    for name in names:
        print(f"Hello, {name.title()}")

greet_all(["alice", "bob"])
```
