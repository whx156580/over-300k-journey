> 本 Notebook 展示了如何利用 Python 的标准流进行交互，并演示模块入口保护的最佳实践。

## 1. 结构化输出
利用 `print` 参数控制输出样式。

```python
print("Item", "Qty", "Price", sep=" | ")
print("-" * 20)
print("Apple", 10, 2.5, sep=" | ")
```

## 2. 安全的输入转换
模拟从 `input()` 获取字符串并安全转换为整数。

```python
def safe_int(val_str):
    try:
        return int(val_str.strip())
    except ValueError:
        return 0

print(f"Parsed: {safe_int('  42  ')}")
print(f"Invalid: {safe_int('abc')}")
```

## 3. 入口保护演示
解释 `__name__` 变量的作用。

```python
print(f"Current module name: {__name__}")
if __name__ == "__main__":
    print("Running as a script!")
else:
    print("I am being imported.")
```
