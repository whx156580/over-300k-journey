# 切片、解包与高频内置函数实战

> 本 Notebook 展示了如何高效重组数据序列，以及如何利用 `zip`, `enumerate`, `any/all` 编写简洁代码。

## 1. 扩展解包 (Extended Unpacking)
利用星号语法捕获序列中的剩余部分。

```python
data = ["2026-01-01", "INFO", "User login", "OK", "0.1s"]
date, level, *details, status, duration = data

print(f"Date: {date}")
print(f"Details: {details}")
print(f"Status: {status}")
```

## 2. 多序列同步遍历
组合使用 `zip` 与 `enumerate`。

```python
users = ["Alice", "Bob", "Charlie"]
roles = ["Admin", "Editor", "Viewer"]

for i, (name, role) in enumerate(zip(users, roles), 1):
    print(f"{i}. {name} is {role}")
```

## 3. 快速条件判定
利用 `any()` 和 `all()` 检查测试状态。

```python
results = [True, True, False, True]
print(f"All passed? {all(results)}")
print(f"Any failed? {not all(results)}")
print(f"Has any success? {any(results)}")
```
