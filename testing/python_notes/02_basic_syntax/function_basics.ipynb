# 函数进阶与作用域 (Functions & Scopes)

> 本 Notebook 展示了 Python 的 LEGB 查找规则、闭包状态保持以及如何避免可变默认参数陷阱。

## 1. 闭包与状态保持
演示如何不使用类就实现一个有“记忆”的函数。

```python
def make_multiplier(factor):
    def multiplier(x):
        return x * factor
    return multiplier

double = make_multiplier(2)
triple = make_multiplier(3)

print(f"2 * 10 = {double(10)}")
print(f"3 * 10 = {triple(10)}")
```

## 2. nonlocal 关键字
在嵌套函数中修改外部变量绑定。

```python
def counter():
    count = 0
    def inc():
        nonlocal count
        count += 1
        return count
    return inc

c = counter()
print(c(), c(), c())
```

## 3. 默认参数陷阱
对比错误用法与最佳实践。

```python
def bad_f(a=[]):
    a.append(1)
    return a

def good_f(a=None):
    if a is None: a = []
    a.append(1)
    return a

print(f"Bad: {bad_f()}, {bad_f()}")
print(f"Good: {good_f()}, {good_f()}")
```
