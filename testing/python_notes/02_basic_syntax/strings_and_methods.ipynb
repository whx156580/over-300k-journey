## 1. f-string 高级格式化
演示对齐、宽度填充与数字精度控制。

```python
data = [("API_Login", 200, 0.456), ("UI_Cart", 404, 2.1)]
print(f"{'Name':<12} | {'Code':>5} | {'Time':>6}")
print("-" * 30)
for name, code, time in data:
    print(f"{name:<12} | {code:>5} | {time:>5.2f}s")
```

## 2. Raw String 与路径
在处理 Windows 文件系统时必用的 `r""` 语法。

```python
path = r"C:\new_folder\test.py"
print(f"Raw Path: {path}")
print(f"Parts: {path.split('\\')}")
```

## 3. 性能测试：Join vs Plus
演示为什么在大规模拼接时必须使用 `.join()`。

```python
import timeit
n = 10000
parts = ["word"] * n

def plus():
    s = ""
    for p in parts: s += p
    return s

def join():
    return "".join(parts)

t1 = timeit.timeit(plus, number=10)
t2 = timeit.timeit(join, number=10)
print(f"'+' time: {t1:.4f}s")
print(f"'join' time: {t2:.4f}s")
```
