## 1. 费用数据聚合
演示如何利用字典进行分类统计。

```python
data = [
    {"cat": "Food", "amt": 20},
    {"cat": "Food", "amt": 15},
    {"cat": "Transport", "amt": 50}
]

stats = {}
for item in data:
    c = item['cat']
    stats[c] = stats.get(c, 0) + item['amt']

print(f"Total by category: {stats}")
```

## 2. 文本规范化预览
使用推导式快速生成转换预览。

```python
raw_files = ["  Report_A.txt", "data 01.csv  "]
preview = [(f, f.strip().lower().replace(" ", "_")) for f in raw_files]
for old, new in preview:
    print(f"{old} -> {new}")
```

## 3. 自动化评分逻辑
比较两个字典并计算得分率。

```python
keys = {1: "A", 2: "B", 3: "C"}
student = {1: "A", 2: "C", 3: "C"}

score = sum(1 for q_id, ans in keys.items() if student.get(q_id) == ans)
print(f"Final Score: {score}/{len(keys)}")
```
