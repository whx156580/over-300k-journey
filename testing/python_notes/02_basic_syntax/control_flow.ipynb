> 本 Notebook 展示了 Python 中的短路逻辑、`for-else` 搜索模式以及 Python 3.10+ 的 `match-case`。

## 1. `for-else` 搜索成功检测
演示如何优雅地处理“查找不到目标”的情况。

```python
items = [1, 3, 5, 7]
target = 4

for item in items:
    if item == target:
        print(f"Found {target}!")
        break
else:
    # 只有当循环由于不满足条件结束（未触发 break）时执行
    print(f"{target} not in list.")
```
