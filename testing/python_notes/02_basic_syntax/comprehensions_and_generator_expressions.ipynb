# 推导式与生成器表达式 (Comprehensions & Generator Expressions)

> 本 Notebook 展示了如何使用 Python 的推导式语法进行高效的数据处理和内存优化。

## 1. 列表推导式：过滤与转换
将复杂循环简化为单行表达式。

```python
data = range(10)
evens_squared = [x**2 for x in data if x % 2 == 0]
print(f"Evens squared: {evens_squared}")
```

## 2. 嵌套推导式：拍平二维结构
处理嵌套列表的常用技巧。

```python
nested = [[1, 2], [3, 4], [5]]
flat = [item for sublist in nested for item in sublist]
print(f"Flattened: {flat}")
```

## 3. 生成器表达式：千万级数据的救星
对比立即求值与惰性求值的内存差异。

```python
import sys
n = 1_000_000
l = [x for x in range(n)]
g = (x for x in range(n))
print(f"List: {sys.getsizeof(l)/1024/1024:.2f} MB")
print(f"Generator: {sys.getsizeof(g)} Bytes")
```
